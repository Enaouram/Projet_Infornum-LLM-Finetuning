```python
import csv
import openai
import sys
import os
from openai import AzureOpenAI
# Increase CSV field size limit to accommodate large fields
csv.field_size_limit(sys.maxsize)

# Configure OpenAI API
client = AzureOpenAI(
    api_key = "xxxxxxxxxxxxxxxxxxxxxxx",
    api_version="xxxx-xx-xx",
    azure_endpoint="https://xxxxxxxxxxxxxx.com/",
    azure_deployment='xxxxxxxxx'
)

def generate_annotation_for_script(script):
    try:
        response = client.chat.completions.create(
            model="gpt-35-turbo",
            messages= [{"role": "user", "content": "Provide an instruction that I can feed to an LLM model so as to match the following SAS programming language script:"+str({script})}],
            temperature=0.5,
            max_tokens=300,  # Adjust based on how lengthy we expect the responses to be
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
        return response.choices[0].message.content
    except Exception as e:
        print(f"An error occurred: {e}")
        return "Annotation could not be generated."


```

```python
# Initialize batch processing parameters
batch_size = 2000
total_rows = len(df)
current_batch = 0  # We'll update this in subsequent cells
```

```python
# Obtenir l'indice de la colonne 'content'
index_col_content = df.columns.get_loc('content')

# Afficher les indices
print(f"L'indice de la colonne 'content' est : {index_col_content}")
```

```python
df['Annotation'] = pd.NA
df
```

```python
index_col_annotation = df.columns.get_loc('Annotation')
# Afficher les indices
print(f"L'indice de la colonne 'Annotation' est : {index_col_annotation}")
```

```python
# Calculate start and end indices for the current batch
start_index = current_batch * batch_size
end_index = min(start_index + batch_size, total_rows)

# Utiliser .iloc pour assigner les annotations
# Notez qu'on doit transformer la sÃ©rie retournÃ©e par apply() en liste pour Ã©viter les problÃ¨mes d'alignement d'index
df.iloc[start_index:end_index, index_col_annotation] = df.iloc[start_index:end_index]['content'].apply(generate_annotation_for_script).tolist()

# Update current_batch for the next run
current_batch += 1

# Let's print some information
print(f"Processed batch {current_batch} (rows {start_index} to {end_index})")
```

```python
import csv
import google.generativeai as genai

# Configure GenerativeAI API
genai.configure(api_key="xxxxxxxxxxxxxxxxxxxxxxxxxxxx")

# Generation configuration (Adjust these if needed)
generation_config = {
    "temperature": 0.7,  # Adjust for creativity
    "top_p": 0.9,  # Adjust for diversity in responses
    "top_k": 40,  # Adjust for controlling the set from which the next token is sampled
    "max_output_tokens": 300  # Adjust according to the expected length of instructions
}

# Initialize GenerativeModel with adjusted configuration for SAS script annotation
model = genai.GenerativeModel(
    model_name="gemini-1.0-pro", generation_config=generation_config
)

# Function to generate annotation (instruction) for SAS script
def generate_annotation_for_script(script):
    prompt = "Provide only the instructions that I can feed to an LLM model so as to obtain the following SAS script:"
    contents = [prompt + " " + script]
    response = model.generate_content(contents=contents)
    if response.parts:
        return response.parts[0].text  # Assuming the first part of the response contains the annotation
    return "Annotation could not be generated."
```
